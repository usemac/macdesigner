```python
from macdesigner.macdesigner import spm_designer
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.ioff()
```

### Designer plane

```python
spm = spm_designer(0.175,0.11)
spm.g = 1e-3
spm.x = 0.6 # design factor  x = (R_r_int + l_m)/R_s_ext
spm.b = 4.5 # design factor  b = l_m/g
spm.beta = 1 # magnet rounding (1: no rounding) 
spm.design()
spm.report_pellegrino()
spm.plot_xb_plane()
spm.save2pyleecan('dicorca_spm_2.json')
```
